```python
# NER --- >

# Named Entity Recognition (NER) is an NLP technique used to find important names and information in
# text-such as people, places, organizations, dates, and amounts-and Label them with their correct
# type.

# In short:
# NER helps a machine understand "who", "what", and "where" from text.

# ExampLe (Very Simple)
# Text:
# "Rohit Sharma Lives in Mumbai.M

# NER Output:
# Rohit Sharma PERSON
# Mumbai + LOCATION (GPE)

# Why Do We Use NER?
# We use NER to convert unstructured text into meaningful, structured data.
# Main Reasons:
# Automatically extract important information
# Reduce manual work in reading documents
# Help machines understand context, not just keywords
# Make text usable for analytics, ML models, and databases

```

```python
import spacy

# Load EngLish NLP model
nlp = spacy.load("en_core_web_sm")
```

```python

# eX-1
text = "Barack Obama was born in Hawaii and became President of the United States."

doc = nlp(text)

for ent in doc.ents:
    print(ent.text, "->", ent.label_)
```

```python
# Ex-2
text = " Google CEO Sundar Pichai announced a 10 billion investment in India on 15th July 2020"


doc = nlp(text)

for ent in doc.ents:
    print(ent.text, "->", ent.label_)
```

```python
# Ex-3: Person & Location
text = "Sachin Tendulkar was born in Mumbai."
doc = nlp(text)
for ent in doc.ents:
    print(ent.text, "->", ent.label_)
```

```python
# ExampLe-4: Organization & Date

text = "Microsoft was founded in 1975 by Bill Gates."

doc = nlp(text)

for ent in doc.ents:
    print(ent.text, "->", ent.label_)
```
