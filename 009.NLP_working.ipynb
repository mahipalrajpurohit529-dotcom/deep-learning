```python
# TF(Term Frequencey)
# 2. IDF(Inverse Document Frequencey)

# TF(t,d) = (Number of occurences of term t in document d)/
# (Total number of terms in the document d)
# example == >
# d1 = people watch cricket .

# tf(people,d1) = 1/3

# d2 = match was match.
# tf(match , d2) = 2/3
```

```python
# IDF(t) = Loge(Total number of documents in the corpus)/
# (Number of documents with term t in them)
# here corpus means number of sentences

# Example == >
# d1 = people watch cricket .
# d2 = cricket was cricket .
# d3 = people write comment .
# d4 = cricket write comment .

# idf(cricket) = Log(4/4)
# idf(people) = Log(4/2)

# we can use this by apply feature_extraction method in sklearn
```

```python
from sklearn. feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer()
tfidf.fit_transform(df['text']).toarray()
```
