```python


# Workflow of Deep Learning

# # Data Ingestion ---- > Data CLeaning and Preprocessing --- >
# Model Sequnetial Architecture(
# Inpput_Layer(Neurons, Activation fn, input_dimension),
# hidden_Layer(Neurons, Activatin_fn),
# output_Layer(Neurons, Activation_fn)
# )
# Model.compile(loss_fn , optimizer , metrics)
# Model. training
# Model Performance Evaluation







import pandas as pd
import numpy as np

df = pd.read_csv("/content/Attrition - Attrition.csv")



df = df.dropna()

x = df.drop(columns=['Attrition'],axis=1)
y = df['Attrition']



categorical_features = ['BusinessTravel',
'Department',
'EducationField',
'Gender',
'JobRole',
'MaritalStatus',
'Over18',
'OverTime']

x_encoded = pd.get_dummies(x,columns= categorical_features ,drop_first=True)

x_encoded = x_encoded.astype(int)



# x_encoded

from sklearn.preprocessing import LabelEncoder

lb = LabelEncoder()

y_encoded = lb.fit_transform(y)

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

x_train ,x_test ,y_train ,y_test = train_test_split(x_encoded,y_encoded,test_size=0.2,random_state=42)



sc = StandardScaler()

x_train_scaled = sc.fit_transform(x_train)
x_test_scaled = sc.transform(x_test)



# now we will create architecture for our model

import tensorflow as tf

model = tf. keras. models. Sequential ([
                tf.keras. layers. Dense(units = 128 , activation = 'relu' ,
                                      input_dim = x_train_scaled. shape[1]),  #input Layer
                tf.keras.layers.Dropout(0.3),
                tf.keras.layers.Dense(units = 64 , activation = 'relu'),      #hidden layer
                tf.keras.layers.Dropout(0.3),
                tf.keras. layers. Dense(units = 1 , activation = 'sigmoid')   #output Layer
])


# Compile the Model 

loss_fn = tf.keras.losses.BinaryCrossentropy()
metrics = ['accuracy']

learning_rate = 0.001
momentum = 0.9
optimizer = tf.keras.optimizers. SGD(learning_rate = learning_rate,
momentum = momentum , nesterov = True)


model.compile(optimizer= optimizer, loss = loss_fn, metrics= metrics)

# Train the Model

model.fit(x_train_scaled,
          y_train,
          epochs = 50,
          batch_size = 8,
          validation_split = 0.1
)

from sklearn.metrics import  accuracy_score

loss, accuracy = model.evaluate(x_test_scaled, y_test)

print(f"Test Loss = {loss:4f}, Test Accuracy = {accuracy:4f}")

```
