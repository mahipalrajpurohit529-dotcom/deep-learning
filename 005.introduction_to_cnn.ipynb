```python
#  Working of CNN --- >



# CNN is used in various applications such as image classification , object detection , face and pattern recognition

#  It is specially designed to process grid like data such as images

#  unlike treditional ANN , CNN automatecally extracts important features like ages , texture , curves ,
#  shapes and objects from images without requiring manual feature engineering

# CNN works by learning special relationships between pixels and gradually building high level features through multiple layers .





# The basic architecture of CNN contains several layers including the input layer , convolutional layer
#  , activation functionn , pooling layer , flatten layer and dense layer

# The input layer recieves the data in the form of pixels 

# after that the convolution layer performs the main feature extraction opperation 
# in this layer , small metrix called filters or kernels slide over the images and
# perform methamatical operations to detect important patters  such as 
# vertical edges , horizontal edges , textures and corners ,etc

# The output generated after convolution is called a feature map 






#  A kernel or filter is generally a small matrix such as 3x3 , 5x5 or 7x7 that moves across the input image 
#  during convolution the kernel values are multiplied with image pixel values and the sum of results forms the feature map

#  small kernels are mostlyused because they capture fine details , require fewer parameters ,
#  reduce computaional cost and help create deeper networks efficiently

#  Larger kernels like 5x5 or 7x7  are used when broader feature extraction is required 
#  but they increase computation and training time 

#  The output size after convolution is calculated using the formula :
#  w-k+2p/s + 1
# where w is the input size , k is the kernel size , p is the padding and s is the stride 








#  stride refers to the number of pixels by which the kernel moves across the image 
# if stride is 1 , then the kernel moves one pixel at a time and captures more detaild features
# if stride is 2 , then the kernel jumps two pixels reducing computation but possibly losing some information 

#  smaller strides are prefered when detailed feature extraction is needed 
#  while larger strides are used to speed up trainin and reduce dimensionality
#  so in most practical CNN architectures , a stride of size 1 is commonly used because it preserves more information from the image 








# Padding is another important concept in CNN
# During convolution , image borders may lose information because the kernels cannot fully fit around the edge pixels 
# padding solves this problem by adding extra pixels around the image boundaries 

# The two common types are valid padding and same padding 
# valid padding means no extra pixels are added while
# the same padding preserves the original image dimensions after convolution 

# Padding helps retain edge information and prevents excessive reduction in image size after multiple convolution operations
```
