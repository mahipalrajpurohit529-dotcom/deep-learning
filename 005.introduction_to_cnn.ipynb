```python

# Types of Layers ---- >
# (1). Dense Layer --- > It is fully connected layer and output of first neuron will be input of
# second neuron .
# (2). Convolutional Layer --- > It is partially connected layer and it detects only specific parts .
# (3). Flatten Layer --- > It will cnvert high-dimensional image to one-dimensional matrix .
# (4). Pooling layer --- > It is Regularization technique , using this method , we can reduce overfitting.
```

```python
# Image --- > Pixels --- > Range(0-255)
# Opx -- > Image Black
# 255px --- > Image White

# And when we perform Normalization , then its range change to 0-1.
```
